```python
# 위 정보로 전기차/인구밀도/자동차 등록댓수 관계를 나타내는 그래프 그릴 것임

import platform

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from matplotlib import font_manager, rc
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname="C:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
else:
    print("It's unknown system. Hangul fonts are not supported!")
    
# plt.rcParams['axes.unicode_minus'] = False
plt.rcParams["figure.figsize"] = [12,6]

%matplotlib inline
```

```python
#'구'로 index 정해주기 
df.set_index('구', inplace=True)
df.head()
```

```python
# 전기자동차 대수 bar graph 만들어보기

plt.figure(figsize=(20,10))

sns.set_color_codes("pastel")
sns.barplot(y=df.index, x=df['전기차수'], color="g", label='각 구 별 전기자동차 등록 대수')

plt.legend(ncol=2, loc=1, fontsize=15, frameon=True)
plt.title('\n 각 구 별 전기자동차 등록 수 현황 \n', fontsize=20)

plt.show()

#강남구가 압도적으로 높음 
```

```python
# 전체 자동차 등록댓수 대비 전기자동차 대수 -> 숫자가 너무 미세해서 안보임 
plt.figure(figsize=(20,10))

sns.set_color_codes("pastel")
sns.barplot(y=df.index, x=df['등록자동차합'], color="g", label='자동차 등록 대수')

sns.set_color_codes("muted")
sns.barplot(y=df.index, x=df['전기차수'], color="g", label='전기자동차 등록 대수')

plt.legend(ncol=2, loc=1, fontsize=15, frameon=True)
plt.title('\n 전체 자동차 등록 수 대비 전기자동차 등록 수 현황 \n', fontsize=20)

plt.show()
```

```python
#전기 자동차 수에 대한  파이 그려보기 
df_pbox = df['전기차수']
explode = [ 0.2 if df_pbox[i]==df_pbox.max() else 0 for i in range(len(df)) ]

df_pbox.plot(kind='pie', 
             title='각 구별 전기자동차 등록 비율 ',
             figsize=(20, 15), 
             fontsize=12, 
             autopct='%1.1f', 
             shadow=False, 
             startangle=270,
             explode = explode,
            )
plt.axis('equal')
plt.legend(df.index[:-5], loc=1)
plt.show()
```

```python
#Scatter로도 표현!
plt.figure(figsize=(10, 10))
plt.scatter(df['인구총합'], df['전기차수'], s=50)
plt.xlabel('인구수')
plt.ylabel('전기자동차수')
plt.grid()
plt.show()
```

```python
fp1 = np.polyfit(df['인구총합'], df['전기차수'], 1)
fp1
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df['인구총합'], df['전기차수'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('인구수')
plt.ylabel('전기자동차')
plt.grid()
plt.show()
```

```python
# 그래프에 의미를 부여해보기 
fp1 = np.polyfit(df['인구총합'], df['전기차수'], 1)

f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)

df['오차'] = np.abs(df['전기차수'] - f1(df['인구총합']))

df_sort = df.sort_values(by='오차', ascending=False)
df_sort.head(5)
```

```python
plt.figure(figsize=(20,20))
plt.scatter(df['인구총합'], df['전기차수'], 
            c=df['오차'], s=30)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

for n in range(10):
    plt.text(df_sort['인구총합'][n]*1.02, df_sort['전기차수'][n]*0.98, 
             df_sort.index[n], fontsize=15)
    
plt.xlabel('인구수')
plt.ylabel('인구당 전기자동차 등록 비율')
plt.colorbar()
plt.grid()
plt.show()
```
